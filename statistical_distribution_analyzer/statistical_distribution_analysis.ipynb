# Statistical Distribution Analysis using Python and NumPy

This script is a demonstration of how to generate and analyze different types of statistical distributions. It makes use of the `numpy` library to generate data following specific distributions, and an assumed `DistributionAnalyzer` class to analyze each dataset.

Each block of code does the following:

1. **Normal Distribution Test:** This block generates a dataset of 1000 points that follow a Normal (or Gaussian) distribution. It then analyzes the distribution using the `DistributionAnalyzer`.

2. **Binomial Distribution Test:** Here, a dataset is generated based on a Binomial distribution. The parameters `n` and `p` are set to 10 and 0.5 respectively. The distribution is then analyzed.

3. **Uniform Distribution Test:** This block creates a dataset with a Uniform distribution and performs the analysis.

4. **Poisson Distribution Test:** A Poisson distributed dataset is created with a lambda value of 10, which is then analyzed.

5. **Exponential Distribution Test:** This creates an Exponential distributed dataset with a scale of 1 and performs the analysis.

6. **Beta Distribution Test:** This block generates a dataset based on a Beta distribution with shape parameters `a` and `b` both set to 2. The distribution is then analyzed.

For each distribution, the `analyze()` method of the `DistributionAnalyzer` class is called. See the class for more information

```python
from src.distribution_analyzer import DistributionAnalyzer
import numpy as np
```

```python
# Normal distribution test
print("Testing Normal distribution")
data_normal = np.random.normal(size=1000)
analyzer_normal = DistributionAnalyzer(data_normal)
analyzer_normal.analyze()

# Binomial distribution test
print("\nTesting Binomial distribution")
data_binomial = np.random.binomial(n=10, p=0.5, size=1000)
analyzer_binomial = DistributionAnalyzer(data_binomial)
analyzer_binomial.analyze()

# Uniform distribution test
print("\nTesting Uniform distribution")
data_uniform = np.random.uniform(size=1000)
analyzer_uniform = DistributionAnalyzer(data_uniform)
analyzer_uniform.analyze()

# Poisson distribution test
print("\nTesting Poisson distribution")
data_poisson = np.random.poisson(lam=10, size=1000)
analyzer_poisson = DistributionAnalyzer(data_poisson)
analyzer_poisson.analyze()

# Exponential distribution test
print("\nTesting Exponential distribution")
data_exponential = np.random.exponential(scale=1, size=1000)
analyzer_exponential = DistributionAnalyzer(data_exponential)
analyzer_exponential.analyze()

# Beta distribution test
print("\nTesting Beta distribution")
data_beta = np.random.beta(a=2, b=2, size=1000)
analyzer_beta = DistributionAnalyzer(data_beta)
analyzer_beta.analyze()
```
